# 이상한 응답 데이터를 찾는다(데이터 범위 : 1~5)

```python
# 숫자형 컬럼 중 최대값이 6 이상인 컬럼만 필터링
max_values = df.select_dtypes('number').max().to_dict()
filtered_max_values = {key: value for key, value in max_values.items() if value >= 6}

print(filtered_max_values)
```

```python
for i in filtered_max_values:
    df=df[df[i]<6]
df
```

# 대소문자 때문에 같은 MBTI라도 다른다고 인식한다

```python
df['MBTI'].unique()
```

# MBTI를 대문자로 바꾼다

```python
df['MBTI']=df['MBTI'].apply(lambda x: str(x).upper())
df
```

# XXYX, XXXY, ABCD 유형의의 MBTI가 없다

```python
df['MBTI'].unique()
```

```python
mbtis = [m1 + m2 + m3 + m4 for m1 in "IE" for m2 in "NS" for m3 in "TF" for m4 in "PJ"]
mbtis
```

# 이상한 MBTI를 지운다

```python
df=df[df['MBTI'].isin(mbtis)]
df
```

```python
questions=pd.read_csv('mbti_questions.csv')
```

# 시각화를 한다

```python
plt.figure(figsize=(100, 50))
num_columns = len(df.select_dtypes('number').columns)
rows = (num_columns // 5) + (num_columns % 5 > 0)
for i, column in enumerate(df.select_dtypes('number').columns):
    tmp=df.drop_duplicates(subset=[column,"MBTI"])
    plt.subplot(rows, 5, i + 1)
    plt.title(questions[questions['QuestionID']==column]['QuestionText'].values[0])
    sns.stripplot(data=tmp, x="MBTI", y=column, jitter=True, size=8)
plt.tight_layout()  # 서브플롯 간의 간격 조정
```
